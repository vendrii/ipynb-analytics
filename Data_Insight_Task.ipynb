# Jawaban Soal 3

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

dessert_sales = {
'Lava Cake': 14,
'Mousse': 5,
'Chocolate Cake': 12,
'Ice Cream': 19,'Truffles': 6,
'Brownie': 8,
'Chocolate Chip Cookie': 12,
'Chocolate Pudding': 9,
'Souffle': 10,
'Chocolate Cheesecake': 17,
'Chocolate Chips': 2,
'Fudge': 9,
'Mochi': 13,
}

data=pd.DataFrame(list(dessert_sales.items()), columns=['X', 'Y'])
menu=data['X']
menu
frek=data['Y']
frek

df = pd.DataFrame ({
    'Menu':menu,
    'Frequency':frek
})

df.sort_values(by='Frequency', inplace=True, ascending=False)

x_coords = np.arange(len(df))

colors = ['#00bbff' for _ in range (len(df))]
colors[-4] = '#6200ff'
colors[-3] = '#6200ff'
colors[-2] = '#6200ff'
colors[-1] = '#6200ff'

plt.bar(x_coords, df['Frequency'], tick_label=df['Menu'], color=colors)
plt.xticks(rotation=90)
plt.ylabel('Order Frequecy')
plt.title('Best Selling Menu')
plt.show()
```

Chart yang digunakan pada case study ke-3 ini adalah bar chart, karena tipe datanya mengandung kategori

---
Berdasarkan data yang ada terdapat 4 menu yang sebaiknya dikeluarkan yakni Brownie, Truffles, Mousse, dan Chocolate Chips. Karena secara frekuensi penjualannya sangat rendah.
Perlunya evaluasi dalam inovasi di ke-4 menu tersebut untuk meningkatkan frekuensi pembelian. Evaluasi yang dapat dilakukan yakni pada Plating nya atau mungkin variansi rasa yang dapat dikolaborasi dengan 4 menu tersebut
